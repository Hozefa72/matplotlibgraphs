```python
batters=pd.read_csv('batter.csv')
```

```python
batters
```

```python
sample=batters.head(100).sample(25,random_state=5)
```

```python
sample
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='k')
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
```

```python
fig,axs=plt.subplots()
axs.scatter(batters['avg'],batters['strike_rate'],color='red',marker='*')
axs.set_title("Batter's Performance")
axs.set_xlabel("Average")
axs.set_ylabel("Strike Rate")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='*')
ax[0].set_title("Avg v/s Strike Rate")
ax[0].set_ylabel("Strike rate")
ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title("Avg v/s Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Average")
```

```python
a=np.arange(0,100,10)
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'],bins=a)
ax[1,1].hist(batters['strike_rate'])
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'])
ax2=fig.add_subplot(2,1,2)
ax2.hist(batters['avg'])
```

```python
plt.figure(figsize=(10,6))
ax=plt.subplot(projection='3d')
ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],color='red',marker='v')
ax.set_title("Batsman's Performance")
ax.set_xlabel("Runs")
ax.set_ylabel("Average")
ax.set_zlabel("Strike Rate")
```

# Surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
xx
```

```python
yy
```

```python
x
```

```python
y
```

```python
z=xx**2+yy**2
z
```

```python
fig=plt.figure(figsize=(10,6))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(20,20))
ax=plt.subplot(projection='3d')
a=ax.plot_surface(xx,yy,z,cmap="ocean")
fig.colorbar(a)
```

# Contour Plot

```python
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
a=ax.contour(xx,yy,z)
fig.colorbar(a)
```
